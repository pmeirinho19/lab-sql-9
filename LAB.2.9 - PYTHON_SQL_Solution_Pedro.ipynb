```python
# 2.Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week(LAB2.5; 8.)

data2 = pd.read_sql_query("SELECT *, WEEKDAY(rental_date) AS weekday_id,CASE  WHEN WEEKDAY(rental_date) = 5 then 'weekend'WHEN WEEKDAY(rental_date) = 6 then 'weekend'ELSE 'workday'END AS day_type from sakila.rental", engine)

data2[['rental_id','rental_date','day_type']].head(10)
```

```python
# 5.How many films include Behind the Scenes content? (LAB2.5; 13.)

data4 = pd.read_sql_query("SELECT * from sakila.film", engine)

data4['special_features'].str.contains('Behind the Scenes')

data4['check_behind_scenes_in'] = data4['special_features'].str.contains('Behind the Scenes')

filter_with_behind_scenes = data4[data4['check_behind_scenes_in']==True]

filter_with_behind_scenes
```

```python
# 6.Which kind of movies (rating) have a mean duration of more than two hours?(LAB2.6;7.)

data5 = pd.read_sql_query("SELECT rating, count( distinct film_id) as films_per_rating, round(avg(length),2)  from sakila.film group by rating", engine)
data5
```
